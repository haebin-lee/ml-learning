스케일링: 데이터의 범위가 달라서 가지는 의미도 각각 다르다. 이걸 맞춰주는 작업 
- 표준화 스케일링 : 평균이 0이 되고, 표준편차가 1이 되도록 데이터를 고르게 분포시키는데 사용 
- 로버스트 스케일링: 데이터에 아웃라이어가 존재하고, 그 영향력을 그대로 유지하고 싶을 때 사용 
- 최소-최대 스케일링: 데이터 분포의 특성을 최대한 그대로 유지하고 싶을 때 사용 
- 정규화 스케일링: 행 기준의 스케일링이 필요할 때 사용하나, 실제로 거의 사용하지 않음

표준화 스케일링 - 데이터에 아웃라이어가 존재할 때 아웃라이어의 영향을 받음, 평균:0 분산:1이 되게끔 분포시키기 때문에 데이터의 기존 분포 형태가 사라지고 정규 분포를 따르는 결과물을 가져온다. 

로버스트 스케일링 - 데이터에 아웃라이어가 존재할 때, 아웃라이어의 영향을 받지 않는다. 변환된 데이터의 범위는 표준화 스케일링이나 최소-최대 스케일링보다 넓게 나타난다. 

최소-최대 스케일링 - 표준화 스케일링과 마찬가지로 아웃라이어의 영향을 받게 된다. 위의 두 스케일러와 비교했을 때 데이터의 기존 분포를 가장 있는 그대로 나타내며 스케일만 변화 시킨다. 데이터는 0-1로 나타난다.

**스케일링 주의 사항**
1. 스케일링 대상에서 종속변수는 제외애햐 한다. 
2. 스케일링 전에 훈련셋과 시험셋을 나누어야 한다. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class', axis=1), data['class'], test_size=0.2, random_state = 100) # 시험셋과 학습셋 분리
```

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X_train)
```

```python
X_train_scaled = mm_scaler.transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

```python
X_train_scaled = mm_scaler.fit_transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

하이퍼 파라미터 튜닝하기
- n_neighbors=5 예측에 가까운 이웃을 몇 개나 고려할지 결정
- wekgnts='uniform' uniform: 모든 포인트에 동일한 가중치 (uniform, distance, 사용자 정의함수)
- metric='minkowski' : 거리 측정 기준 
- n_jobs = None: 실행할 병렬 작업수

```python
knn = KNeighborsClassifier(n_neighbors=7) 
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

```python
scores =[]
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    scores.append(accuracy_score(y_test, pred))
```

```python
sns.lineplot(x=range(1, 21), y=scores)
```

```python
# 대체로 매개변수 값이 클수록 더 나은 예측을 보여주지만 13 이후로는 더 나은 개선이 보이지 않으므로.. 13이 합리적인 선택!
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```
