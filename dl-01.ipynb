### 01. Overview of Deep Learning

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 5, 0.2) # incrase by 0.2 from 0 to 5(not included)
t
```

```python
t2 = t**2
t3 = t**3
print(t2)
print(t3)
```

```python
mu, sigma = 100, 15 # mu: mean, sigma: standard deviation
x = np.random.normal(mu, sigma, 10000)
```

```python
# show 2 plots in 1 figure.
plt.plot(t, t2, label='t2')
plt.plot(t, t3, label='t3')
plt.title('show 2 plots in 1 figure')
plt.legend()
plt.show()
```

```python
# use subplot

plt.subplot(1, 2, 1)
plt.plot(t, t)

plt.subplot(1, 2, 2)
plt.plot(t, t3)
plt.show()
```

```python
# histogram 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram of Normal distribution')
plt.hist(x, 50, density = 1, facecolor='g', alpha=0.75)
plt.show()
```
